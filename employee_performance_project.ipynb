```python
data.corr()
```

```python
y = data.PerformanceRating
# x = data.iloc[:,[4,5,9,16,20,21,22,23,24]] all predictors were selected it resulted in dropping of accuracy.

x = data.iloc[:,[4,5,9,16,20,21,22,23,24]]  # taking only variables with corelation coefficient greater than 0.1
x.head()
```

```python
x_train,x_test,y_train,y_test =  train_test_split(x,y, test_size=0.3,random_state=10)
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
print(sc)
print(x_train)
print(x_test)
```

```python
x_train.shape

```

```python
# Logistic regressin
from sklearn.linear_model import LogisticRegression
model_logr = LogisticRegression()
model_logr.fit(x_train,y_train)
```

```python
# predicting the model
y_predict_log = model_logr.predict(x_test)
```

```python
print(accuracy_score(y_test,y_predict_log))
print(classification_report(y_test,y_predict_log))
```
